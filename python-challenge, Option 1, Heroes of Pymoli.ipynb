```python
import pandas as pd
import numpy as np 
```

```python
df = pd.read_csv('purchase_data.csv')
df.head()
```

```python
#find number of unique players or Player Count
grouped_player_df = df.groupby(['SN'])
grouped_player_df.head()
player_count = grouped_player_df['SN'].nunique()
player_total = player_count.sum()
player_total

player_df = pd.DataFrame({
    'Total Number of Players': [player_total],
})
player_df
```

```python
# Number of Unique Items
num_items = df['Item ID'].nunique()
num_items
# Average Purchase Price
avg_item_price = round(df['Price'].mean(), 2)
avg_item_price
# Total Number of Purchases
num_purch = df['Purchase ID'].nunique()
num_purch
# Total Revenue
total_rev = avg_item_price * num_purch

purchasing_df = pd.DataFrame({
    'Number of Unique Items': [num_items],
    'Average Price': (f'${avg_item_price}'), 
    'Number of Purchases': num_purch, 
    'Total Revenue ': (f'${total_rev}')
})
purchasing_df
# print(num_items, avg_item_price, num_purch, total_rev)
```

```python
# drop duplicate 'SN'set index to player 'SN' 
df_players = df.drop_duplicates(['SN'])
df_players['SN'].duplicated().sum()
df_players = df_players.set_index('SN')
# Percentage and Count of Male Players
male_players_count = df_players.loc[df_players['Gender']=='Male',]
male_players_count=male_players_count.value_counts()
male_players_count=male_players_count.sum()
male_player_per = round((male_players_count/player_total)*100, 2)
# Percentage and Count of Female Players
female_players_count = df_players.loc[df_players['Gender']=='Female',]
female_players_count=female_players_count.value_counts()
female_players_count = female_players_count.sum()
female_players_count
female_player_per = round((female_players_count/player_total)*100, 2)
female_player_per
# Percentage and Count of Other / Non-Disclosed
other_players_count = df_players.loc[df_players['Gender']=='Other / Non-Disclosed',]
other_players_count=other_players_count.value_counts()
other_players_count=other_players_count.sum()
other_players_count
other_player_per = round((other_players_count/player_total)*100, 2)
other_player_per




gen_list = [{'Total Count': male_players_count,
             'Percent of Players': male_player_per},
             {'Total Count': female_players_count,
              'Percent of Players': female_player_per},
             {'Total Count':other_players_count,
            'Percent of Players': other_player_per}
           ]
gen_total_df = pd.DataFrame(gen_list, index=['Male', 'Female', 'Other / Non-Disclosed'])

gen_total_df


```

```python
# print(df.columns)
# Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender
male_players_df = df.loc[df['Gender']=='Male',:]
male_purc_count = male_players_df['Purchase ID'].count()
male_purc_avg = round(male_players_df['Price'].mean(),2)
male_purc_total= round(male_players_df['Price'].sum(),2)
male_total_per = round(male_purc_total/male_players_count, 2)

female_players_df = df.loc[df['Gender']=='Female',:]
female_purc_count = female_players_df['Purchase ID'].count()
female_purc_avg = round(female_players_df['Price'].mean(),2)
female_purc_total= round(female_players_df['Price'].sum(),2)
female_total_per = round(female_purc_total/female_players_count, 2)

other_players_df = df.loc[df['Gender']=='Other / Non-Disclosed',:]
other_purc_count = other_players_df['Purchase ID'].count()
other_purc_avg = round(other_players_df['Price'].mean(),2)
other_purc_total= round(other_players_df['Price'].sum(),2)
other_total_per = round(other_purc_total/other_players_count, 2)

# Create a summary data frame to hold the results

gen_purc = [{'Purchase Count': male_purc_count,
             'Average Purchase Size': male_purc_avg,
            'Total Purchase Value': male_purc_total,
            'Avg Total Purchase per Person': male_total_per},
            {'Purchase Count': female_purc_count,
             'Average Purchase Size': female_purc_avg,
            'Total Purchase Value': female_purc_total,
            'Avg Total Purchase per Person': female_total_per},
            {'Purchase Count': other_purc_count,
             'Average Purchase Size': other_purc_avg,
            'Total Purchase Value': other_purc_total,
            'Avg Total Purchase per Person': other_total_per}]
# print(gen_purc)

perc_total_df = pd.DataFrame(gen_purc, index=['Male', 'Female', 'Other / Non-Disclosed'])

perc_total_df
# Optional: give the displayed data cleaner formatting

# Display the summary data frame

```
